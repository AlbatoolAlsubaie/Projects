## Data Wrangling: Change data format to datetime in the date columns

```python
from datetime import datetime
```

```python
df['ScheduledDay'] = pd.to_datetime (df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime (df['AppointmentDay'])
```

```python
df['ScheduledDay'] = df['ScheduledDay'].dt.to_period('D')
df['AppointmentDay'] = df['AppointmentDay'].dt.to_period('D')
```

```python
import numpy as np
```

```python
#I added a column called “Date Difference” where it calculates the difference between the date of booking and the date of appointment

df['Date_Difference'] = (df['AppointmentDay'] - df['ScheduledDay']) / np.timedelta64(1, 'D')
```

## Q2: Is there a correlation between the date of booking the appointment and patients arrival?

```python
#Calculating the arrival percentage for patients.

Percentage = df.value_counts(["No-show", "Date_Difference"], normalize=True).reset_index()
```

```python
Percentage
```

```python
#I calculated the percentage of the patient arrival for each date difference, to see how many patients who booked the appointment on the same day showed up to the appointment vs days later, 
#I named it “Arrival Percentage”.

Percentage = Percentage.rename(columns={0:'Arrival_Percentage'})
```

```python
# used seaborn to make a line plot showing the desired data.

import seaborn as sns
figure(figsize=(14, 10), dpi=90)
sns.lineplot(x="Date_Difference", y="Arrival_Percentage", hue="No-show", data=Percentage, marker="o").set(title= "Patients Arrival to Appointment by Date of Booking")
```

From the plot above, we can see that approximately 33% of the patients who arrived to their appointments have booked the appointment on the same day, which is the largest percentage of the patients in this data. This means that if a patient books their appointment in the same day they are more likely to show up.
